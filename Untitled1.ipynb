```python
# StwÃ³rz przykÅ‚adowe dane samochodÃ³w
import pandas as pd
import numpy as np

# Tworzymy 50 samochodÃ³w
np.random.seed(42)  # dla powtarzalnoÅ›ci

dane = {
    'id': range(1, 51),
    'marka': np.random.choice(['Toyota', 'BMW', 'Audi', 'Mercedes', 'Ford', 'Opel', 'Volkswagen'], 50),
    'model': ['Model_' + str(i) for i in range(1, 51)],
    'rok_produkcji': np.random.randint(2010, 2023, 50),
    'przebieg_km': np.random.randint(10000, 200000, 50),
    'paliwo': np.random.choice(['benzyna', 'diesel', 'hybryda', 'elektryk'], 50),
    'skrzynia': np.random.choice(['manualna', 'automatyczna'], 50),
    'moc_km': np.random.randint(80, 300, 50),
    'cena_pln': np.random.randint(20000, 150000, 50)
}

# StwÃ³rz DataFrame
df_samochody = pd.DataFrame(dane)

# Dodaj trochÄ™ brakujÄ…cych danych (Å¼eby byÅ‚o co czyÅ›ciÄ‡)
for i in range(10):
    row = np.random.randint(0, 50)
    col = np.random.choice(['przebieg_km', 'moc_km', 'paliwo'])
    df_samochody.loc[row, col] = np.nan

# Zapisz do pliku
df_samochody.to_csv('data/samochody.csv', index=False, encoding='utf-8')

print(f"âœ… Utworzono plik: data/samochody.csv")
print(f"   Wymiary: {df_samochody.shape[0]} wierszy, {df_samochody.shape[1]} kolumn")
print("\nğŸ“‹ PodglÄ…d danych:")
print(df_samochody.head())
```
