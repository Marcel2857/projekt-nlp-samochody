```python
# Stw√≥rz naprawiony main_clean.py
clean_code_fixed = '''"""
main_clean.py - Projekt NLP: Przewidywanie cen samochod√≥w
Autor: [Twoje Imiƒô]
Data: [Data]
Opis: System ML do przewidywania cen samochod√≥w na podstawie ich cech
"""

# ========== IMPORTY ==========
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

def main():
    """G≈Ç√≥wna funkcja programu"""
    
    print("=" * 70)
    print("SYSTEM PRZEWIDYWANIA CEN SAMOCHOD√ìW")
    print("=" * 70)
    
    # ========== 1. WCZYTANIE DANYCH ==========
    print("\\n1. WCZYTANIE DANYCH")
    print("-" * 40)
    
    sciezka = 'data/cars93_real.csv'
    
    if not os.path.exists(sciezka):
        print(f"B≈ÇƒÖd: Brak pliku {sciezka}")
        print("Pobierz z: https://raw.githubusercontent.com/selva86/datasets/master/Cars93.csv")
        return
    
    # Wczytaj dane
    data = pd.read_csv(sciezka)
    print(f"Wczytano: {len(data)} rekord√≥w")
    print(f"Liczba cech: {len(data.columns)}")
    
    # ========== 2. ANALIZA DANYCH ==========
    print("\\n2. ANALIZA DANYCH")
    print("-" * 40)
    
    # Podstawowe statystyki
    if 'Price' in data.columns:
        print("Analiza zmiennej docelowej (Price):")
        print(f"  ≈örednia: ${data['Price'].mean():.2f}")
        print(f"  Mediana: ${data['Price'].median():.2f}")
        print(f"  Min: ${data['Price'].min():.2f}")
        print(f"  Max: ${data['Price'].max():.2f}")
        print(f"  Odchylenie standardowe: ${data['Price'].std():.2f}")
    
    # Sprawd≈∫ brakujƒÖce dane
    missing = data.isnull().sum().sum()
    if missing > 0:
        print(f"\\nUwaga: Znaleziono {missing} brakujƒÖcych warto≈õci")
    
    # ========== 3. PRZYGOTOWANIE DANYCH ==========
    print("\\n3. PRZYGOTOWANIE DANYCH")
    print("-" * 40)
    
    # Wybierz tylko kolumny numeryczne
    numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
    
    if 'Price' not in numeric_cols:
        print("B≈ÇƒÖd: Brak kolumny 'Price' w danych numerycznych")
        return
    
    # Utw√≥rz DataFrame z tylko numerycznymi kolumnami
    df_numeric = data[numeric_cols].copy()
    
    # Uzupe≈Çnij brakujƒÖce warto≈õci
    df_numeric = df_numeric.fillna(df_numeric.mean())
    
    # Przygotuj dane do modelowania
    X = df_numeric.drop(columns=['Price'])
    y = df_numeric['Price']
    
    print(f"Liczba cech: {X.shape[1]}")
    print(f"Liczba pr√≥bek: {X.shape[0]}")
    
    # Normalizacja danych
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    print("Dane znormalizowane (StandardScaler)")
    
    # ========== 4. PODZIA≈Å DANYCH ==========
    print("\\n4. PODZIA≈Å DANYCH")
    print("-" * 40)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.25, random_state=42
    )
    
    print(f"Zbi√≥r treningowy: {X_train.shape[0]} pr√≥bek")
    print(f"Zbi√≥r testowy: {X_test.shape[0]} pr√≥bek")
    
    # ========== 5. TRENOWANIE MODELU ==========
    print("\\n5. TRENOWANIE MODELU")
    print("-" * 40)
    
    # Model Random Forest
    print("Model: Random Forest Regressor")
    
    # Domy≈õlny model
    model_default = RandomForestRegressor(
        n_estimators=100,
        max_depth=10,
        random_state=42
    )
    
    model_default.fit(X_train, y_train)
    y_pred_default = model_default.predict(X_test)
    
    mae_default = mean_absolute_error(y_test, y_pred_default)
    r2_default = r2_score(y_test, y_pred_default)
    
    print(f"\\nWyniki modelu domy≈õlnego:")
    print(f"  MAE: ${mae_default:.2f}")
    print(f"  R¬≤: {r2_default:.3f}")
    
    # ========== 6. FINE-TUNING ==========
    print("\\n6. DOSTRAJANIE PARAMETR√ìW (FINE-TUNING)")
    print("-" * 40)
    
    try:
        # Definiuj przestrze≈Ñ parametr√≥w
        param_grid = {
            'n_estimators': [50, 100, 150],
            'max_depth': [5, 10, 15, None],
            'min_samples_split': [2, 5, 10]
        }
        
        print("Przeszukiwanie siatki parametr√≥w...")
        
        grid_search = GridSearchCV(
            RandomForestRegressor(random_state=42),
            param_grid,
            cv=5,
            scoring='neg_mean_absolute_error',
            n_jobs=-1,
            verbose=0
        )
        
        grid_search.fit(X_train, y_train)
        
        print(f"\\nNajlepsze parametry znalezione:")
        for param, value in grid_search.best_params_.items():
            print(f"  {param}: {value}")
        
        # U≈ºyj najlepszego modelu
        best_model = grid_search.best_estimator_
        y_pred_best = best_model.predict(X_test)
        
        mae_best = mean_absolute_error(y_test, y_pred_best)
        r2_best = r2_score(y_test, y_pred_best)
        
        print(f"\\nWyniki po dostrojeniu:")
        print(f"  MAE: ${mae_best:.2f}")
        print(f"  R¬≤: {r2_best:.3f}")
        
        if mae_best < mae_default:
            improvement = (mae_default - mae_best) / mae_default * 100
            print(f"  Poprawa: {improvement:.1f}%")
        
        model = best_model
        y_pred = y_pred_best
        mae_final = mae_best
        r2_final = r2_best
        
    except Exception as e:
        print(f"Uwaga: Fine-tuning nie powi√≥d≈Ç siƒô: {e}")
        print("U≈ºywam modelu domy≈õlnego")
        model = model_default
        y_pred = y_pred_default
        mae_final = mae_default
        r2_final = r2_default
    
    # ========== 7. ANALIZA WA≈ªNO≈öCI CECH ==========
    print("\\n7. ANALIZA WA≈ªNO≈öCI CECH")
    print("-" * 40)
    
    if hasattr(model, 'feature_importances_'):
        # Pobierz wa≈ºno≈õƒá cech
        feature_importance = pd.DataFrame({
            'feature': X.columns,
            'importance': model.feature_importances_
        }).sort_values('importance', ascending=False)
        
        print("Najwa≈ºniejsze cechy wp≈ÇywajƒÖce na cenƒô:")
        for idx, row in feature_importance.head(5).iterrows():
            print(f"  {row['feature']}: {row['importance']:.3f}")
    else:
        print("Model nie dostarcza informacji o wa≈ºno≈õci cech")
    
    # ========== 8. WIZUALIZACJA ==========
    print("\\n8. WIZUALIZACJA WYNIK√ìW")
    print("-" * 40)
    
    # Utw√≥rz folder na wykresy
    os.makedirs('plots', exist_ok=True)
    
    # Wykres 1: Por√≥wnanie rzeczywistych i przewidywanych warto≈õci
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.scatter(y_test, y_pred, alpha=0.6, edgecolors='k', linewidth=0.5)
    
    # Linia idealnego dopasowania
    min_val = min(y_test.min(), y_pred.min())
    max_val = max(y_test.max(), y_pred.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2)
    
    plt.xlabel('Rzeczywista cena ($)')
    plt.ylabel('Przewidziana cena ($)')
    plt.title('Rzeczywiste vs Przewidziane ceny')
    plt.grid(True, alpha=0.3)
    
    # Wykres 2: Rozk≈Çad b≈Çƒôd√≥w
    plt.subplot(1, 2, 2)
    errors = y_test - y_pred
    plt.hist(errors, bins=15, edgecolor='black', alpha=0.7)
    plt.axvline(x=0, color='red', linestyle='--', linewidth=2)
    plt.xlabel('B≈ÇƒÖd przewidywania ($)')
    plt.ylabel('Liczba pr√≥bek')
    plt.title('Rozk≈Çad b≈Çƒôd√≥w przewidywania')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('plots/results_visualization.png', dpi=120, bbox_inches='tight')
    plt.show()
    
    print("Zapisano wykres: plots/results_visualization.png")
    
    # ========== 9. PRZYK≈ÅAD PRZEWIDYWANIA ==========
    print("\\n9. PRZYK≈ÅAD PRZEWIDYWANIA")
    print("-" * 40)
    
    if len(X_test) > 0:
        # Wybierz losowƒÖ pr√≥bkƒô testowƒÖ
        sample_idx = 0
        sample_features = X_test[sample_idx].reshape(1, -1)
        actual_price = y_test.iloc[sample_idx]
        predicted_price = model.predict(sample_features)[0]
        
        # Znajd≈∫ oryginalny rekord
        original_record = data.iloc[data.index[y_test.index[sample_idx]]]
        
        print("Przyk≈Çadowy samoch√≥d:")
        if 'Manufacturer' in original_record and 'Model' in original_record:
            print(f"  {original_record['Manufacturer']} {original_record['Model']}")
        if 'Type' in original_record:
            print(f"  Typ: {original_record['Type']}")
        
        print(f"\\nCena rzeczywista: ${actual_price:.2f}")
        print(f"Cena przewidziana: ${predicted_price:.2f}")
        print(f"R√≥≈ºnica: ${abs(actual_price - predicted_price):.2f}")
        
        error_percent = abs(actual_price - predicted_price) / actual_price * 100
        print(f"B≈ÇƒÖd wzglƒôdny: {error_percent:.1f}%")
    
    # ========== 10. PODSUMOWANIE ==========
    print("\\n" + "=" * 70)
    print("PODSUMOWANIE")
    print("=" * 70)
    
    print(f"\\nNajlepszy model: Random Forest Regressor")
    print(f"≈öredni b≈ÇƒÖd bezwzglƒôdny (MAE): ${mae_final:.2f}")
    print(f"Wsp√≥≈Çczynnik determinacji (R¬≤): {r2_final:.3f}")
    
    if r2_final > 0:
        print(f"Model wyja≈õnia {r2_final * 100:.1f}% wariancji danych")
    
    print(f"\\nB≈ÇƒÖd wzglƒôdny: {(mae_final / y.mean()) * 100:.1f}% ≈õredniej ceny")
    
    print("\\n" + "=" * 70)
    print("PROGRAM ZAKO≈ÉCZONY")
    print("=" * 70)

if __name__ == "__main__":
    main()
'''

# Zapisz naprawiony plik
with open('main_clean_fixed.py', 'w', encoding='utf-8') as f:
    f.write(clean_code_fixed)

print("‚úÖ Utworzono: main_clean_fixed.py")
print("Naprawiony kod - bez b≈Çƒôd√≥w w f-string")
print("Uruchom: python main_clean_fixed.py")
```
