```python
import os

# Lista folder√≥w do stworzenia
foldery = [
    'data',          # tu bƒôdƒÖ dane
    'src',           # tu bƒôdƒÖ nasze pliki .py
    'images',        # tu bƒôdƒÖ wykresy
    'notebooks',     # tu bƒôdƒÖ notebooki
    'tests'          # tu bƒôdƒÖ testy
]

print("Tworzƒô strukturƒô folder√≥w...")
for folder in foldery:
    if not os.path.exists(folder):
        os.makedirs(folder)
        print(f"‚úì Utworzono: {folder}/")
    else:
        print(f"‚úì Folder ju≈º istnieje: {folder}/")

print("\n‚úÖ Gotowe! Struktura projektu:")
for folder in foldery:
    print(f"  {folder}/")
```

```python
# Stw√≥rz przyk≈Çadowe dane samochod√≥w
import pandas as pd
import numpy as np

# Tworzymy 50 samochod√≥w
np.random.seed(42)  # dla powtarzalno≈õci

dane = {
    'id': range(1, 51),
    'marka': np.random.choice(['Toyota', 'BMW', 'Audi', 'Mercedes', 'Ford', 'Opel', 'Volkswagen'], 50),
    'model': ['Model_' + str(i) for i in range(1, 51)],
    'rok_produkcji': np.random.randint(2010, 2023, 50),
    'przebieg_km': np.random.randint(10000, 200000, 50),
    'paliwo': np.random.choice(['benzyna', 'diesel', 'hybryda', 'elektryk'], 50),
    'skrzynia': np.random.choice(['manualna', 'automatyczna'], 50),
    'moc_km': np.random.randint(80, 300, 50),
    'cena_pln': np.random.randint(20000, 150000, 50)
}

# Stw√≥rz DataFrame
df_samochody = pd.DataFrame(dane)

# Dodaj trochƒô brakujƒÖcych danych (≈ºeby by≈Ço co czy≈õciƒá)
for i in range(10):
    row = np.random.randint(0, 50)
    col = np.random.choice(['przebieg_km', 'moc_km', 'paliwo'])
    df_samochody.loc[row, col] = np.nan

# Zapisz do pliku
df_samochody.to_csv('data/samochody.csv', index=False, encoding='utf-8')

print(f"‚úÖ Utworzono plik: data/samochody.csv")
print(f"   Wymiary: {df_samochody.shape[0]} wierszy, {df_samochody.shape[1]} kolumn")
print("\nüìã PodglƒÖd danych:")
print(df_samochody.head())
```
